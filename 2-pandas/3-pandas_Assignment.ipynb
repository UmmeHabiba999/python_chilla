### -Prevalence of obesity in the adult population (18 years and older) 

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
```

### -Import dataset from FAO.org 

```python
obesity_df=pd.read_csv('resources/obesity_in_adults.csv')
obesity_df.head()
```

### -change dataset to Excel

### -Basic Statistics

```python
obesity_df.describe()
```

### -Dropping unnecessary columns

```python
obesity_df2 = obesity_df.drop(['Domain','Note','Flag','Item'], axis=1)
obesity_df2
```

```python
obesity_df2.mean()
```

### -Taking mean with Group by (Pivot)

```python
obesity_df2.groupby(['Area','Year','Value']).mean()
```

```python
obesity_df2.value_counts('Year')
```

```python
obesity_df2['Value'].mean()
```

```python
greater_obesity = obesity_df2[obesity_df2['Value']>16]
greater_obesity
```
