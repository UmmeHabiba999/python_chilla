### All the code would be same as lineplot, but change only name from lineplot to barplot in code

```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load data set

flowers = sns.load_dataset("iris")
```

```python
# plot figure size
plt.figure(figsize=(10,6))

# draw plot
sns.barplot(x="species", y="sepal_width" , data= flowers, ci=None)
sns.set_style("dark")
plt.title("Iris flowers plot")

plt.show()
```

#### with different x and y variables

```python
# plot figure size

plt.figure(figsize=(10,6))

# draw plot

sns.barplot(x="species", y="petal_length" , data= flowers,ci=None)
sns.set_style("dark")
plt.title("Iris flowers plot")

plt.show()
```

#### draw bar plot with different data set, Titanic data set thats available within seaborn library

```python
#load data set
titanic_dataset = sns.load_dataset("titanic")
titanic_dataset 
```

```python
# assign figure size
plt.figure(figsize=(10,6))

# draw barplot
sns.barplot(x="who" , y="alone" , hue="sex" , data= titanic_dataset) #hue is used to assign different colors
plt.title("Titanic Graph")
plt.show()
```

```python
# assign figure size
plt.figure(figsize=(10,6))

# draw barplot
sns.barplot(x="sex" , y="alone" , hue="who" , data= titanic_dataset) #hue is used to assign different colors
plt.title("Titanic Graph")
plt.show()
```

#### to change the order, like to put "female" bar first 

```python
# assign figure size
plt.figure(figsize=(10,6))

# draw barplot
sns.barplot(x="sex" , y="alone" , hue="who" , data= titanic_dataset ,order=["female" , "male"] ) 
plt.title("Titanic Graph")
plt.show()
```

#### use different color pallets

```python
# assign figure size
plt.figure(figsize=(10,6))

# draw barplot
sns.barplot(x="sex" , y="alone" , hue="who" , data= titanic_dataset ,order=["female" , "male"] , color="red") 
plt.title("Titanic Graph")
plt.show()
```

#### To remove black lines over the bars/Confidence Interval (ci)

```python
# assign figure size
plt.figure(figsize=(10,6))

# draw barplot
sns.barplot(x="sex" , y="alone" , hue="who" , data= titanic_dataset ,order=["female" , "male"] , color="red" ,ci=None) 
plt.title("Titanic Graph")
plt.show()
```

#### To use different color palettes

```python
# assign figure size
plt.figure(figsize=(8,6))

# draw barplot
sns.barplot(x="sex" , y="alone" , hue="who" , data= titanic_dataset ,order=["female" , "male"] , color="red" ,ci=None ,
           palette="pastel") 

plt.title("Titanic Graph")
plt.show()
```

#### To use the estimator such as Mean, median etc, estimator dosnt work on catagorical values

```python

# for estimator we need to import Numpy 
from numpy import median
# assign figure size
plt.figure(figsize=(10,6))

# draw barplot
sns.barplot(x="class" , y="fare" , hue="who" , data= titanic_dataset, estimator= median) 

plt.title("Titanic Graph")
plt.show()

```

#### For color saturation in plot.. to make bars more bright or dark

```python
# assign figure size
plt.figure(figsize=(10,6))

# draw barplot
sns.barplot(x="class" , y="fare" , hue="who" , data= titanic_dataset, estimator= median , saturation= 0.5) 

plt.title("Titanic Graph")
plt.show()
```

#### Horizontal Plot , If you give a categorical variable to  “y “  , the plot will be horizontal.

```python
# assign figure size
plt.figure(figsize=(8,6))

# draw barplot
sns.barplot(x="fare" , y="class" , hue="who" , data= titanic_dataset, estimator= median , saturation= 0.5) 

plt.title("Titanic Graph")
plt.show()
```

```python
# bars without colors etc
#  linewidth is for borderline width of the bar
# errcolor is for the line/error line on the bar"1" is white. "0" is black, 
# values between 0 and 1 are its diff color like grey
# Edgecolor is Bar edge /outer lines color, 0 and 1 for white and black same as errcolor

# assign figure size
plt.figure(figsize=(8,6))

# draw barplot
sns.barplot(x="class" , y="fare", data= titanic_dataset, linewidth=2 , facecolor= (1,1,1,0) , 
            errcolor= "0" , edgecolor="0") 

plt.title("Titanic Graph")
plt.show()
```
