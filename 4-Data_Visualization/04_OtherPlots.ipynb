## -Line plot with multi facets

```python
# import libraries

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data set

dots=sns.load_dataset("dots")
dots.head()
```

```python
sns.set_theme(style="ticks")
#Defining a color palette
my_palet = sns.color_palette('rocket_r')
#plot lineplot
sns.relplot(data=dots,x="time",y="firing_rate",hue="coherence",size="choice",col="align", kind="line", 
            size_order=["T1","T2"],palette= my_palet, height=5, aspect=0.75,facet_kws=dict(sharex=False),
)
```

#### Different Plots with real data set

```python
chilla = pd.read_csv("chilla_data.csv")
chilla.head()
```

```python
chilla.columns
```

#### Changing Binary optionset data to Numeric data

```python
field = chilla['field_of_study']
vaccinated = chilla['Vaccinated'].map({'Yes': 1, 'No': 0})
gender =chilla['Gender'].map({'Male': 1, 'Female': 0})
glasses = chilla['Glasses'].map({'Yes': 1, 'No': 0})
marital = chilla['Marital Status'].map({'Yes': 1, 'No': 0})
area = chilla['Area'].map({'Urbun': 1, 'Rural': 0})
```

#### Scatter Plot

```python


import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")


# Draw a scatter plot while assigning point colors and sizes to different variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
sns.scatterplot(x="First line of code", y="Height",
                hue=gender, size=vaccinated,
                hue_order=gender, color='red',
                sizes=(1, 8), linewidth=0,
                data=chilla, ax=ax)

```

#### Line Plot

```python
import seaborn as sns
sns.set_theme(style="darkgrid")

chilla = pd.read_csv("chilla_data.csv")

# Plot the responses for different events and regions
sns.lineplot(x="Age", y="Weight in kg",
             hue=area, style=gender,
             data=chilla)
```

#### Displot

```python
import seaborn as sns

sns.displot(
    chilla, x="Age", col=area, row=gender,
    binwidth=3, height=3, facet_kws=dict(margin_titles=True),
)
plt.figure(figsize=(18,6))
```

#### Violinplot

```python


import seaborn as sns
sns.set_theme(style="whitegrid")

# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=chilla, x="Age", y="coding hours", hue=gender,
               split=True, inner="quart", linewidth=1,
               )
sns.despine(left=True)

```

#### lmPlot

```python
import seaborn as sns
chilla = pd.read_csv("chilla_data.csv")

sns.set_theme(style="ticks")



# Show the results of a linear regression within each dataset
sns.lmplot(x="Height", y="Weight in kg", col="Age", hue="Age", data=chilla,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```

#### Jointplot

```python
import numpy as np
import seaborn as sns
sns.set_theme(style="ticks")



sns.jointplot(data=chilla, x="Age", y="Height", color="#4CB391")
```
