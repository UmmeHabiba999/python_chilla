```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#load dataset

iris_data = sns.load_dataset("iris")
iris_data
```

##  -Adding Name of the graph

```python
# draw a line plot

sns.lineplot(x="sepal_length" ,y="sepal_width" , data= iris_data)
plt.title("Iris Flowers Graph")
plt.show()
```

## - Adding x and y limits


```python
# draw a line plot

sns.lineplot(x="sepal_length" ,y="sepal_width" , data= iris_data)
plt.title("Iris Flowers Graph")

plt.xlim(2)
plt.ylim(1)
plt.show()
```

## - set styles

- Dark Grid
- Whitegrid
- Dark
- White
- Ticks

```python
# draw a line plot
sns.lineplot(x="sepal_length" ,y="sepal_width" , data= iris_data)
plt.title("Iris Flowers Graph")
sns.set_style("dark")
# sns.set_style(style=None , rc= None)

plt.show()
```

#### with Different style "white grid"

```python
# draw a line plot
sns.lineplot(x="sepal_length" ,y="sepal_width" , data= iris_data,ci=None)
plt.title("Iris Flowers Graph")
sns.set_style("whitegrid")
# sns.set_style(style=None , rc= None)

plt.show()
```

## - size of figure

```python
#size of figure

plt.figure(figsize=(10,6))

# draw a line plot

sns.lineplot(x="sepal_length" ,y="sepal_width" , data= iris_data)
plt.title("Iris Flowers Graph")
sns.set_style("dark")

plt.show()
```

```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load Dataset

chilla_data = pd.read_csv('chilla_data.csv')
chilla_data.head()
```

### Changing Binary optionset data to Numeric data

```python
field = chilla_data['field_of_study']
marital = chilla_data['Marital Status'].map({'Yes': 1, 'No': 0})
vaccinated = chilla_data['Vaccinated'].map({'Yes': 1, 'No': 0})
gender = chilla_data['Gender'].map({'Male':1 ,'Female':0})
area = chilla_data['Area'].map({'Urbun': 1, 'Rural': 0})
glasses = chilla_data['glasses'].map({'Yes': 1, 'No': 0})
```

#### Line Plot

```python
# select axis

xs = field
ys = chilla_data['coding hours']

plt.figure(figsize= (20,8))
a = sns.lineplot(x= xs,y=ys,data= chilla_data, ci=None)
a.set(xlabel='field of study', ylabel='Coding Hours')

# Add Title of Graph
plt.title('field of study vs coding hours')

plt.show()
```

```python
xs= chilla_data['What are you?']
ys= chilla_data['Purpose_for_chilla']
plt.figure(figsize=(12,6))
# xs, ys = zip(*sorted(zip(xs, ys)))
sns.lineplot(y=xs, x=ys, data =chilla_data )
plt.title('qualification  and purpose')
plt.show()
```
